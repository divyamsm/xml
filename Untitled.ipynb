```python
def kfold(df,n):
    part = 1.00/n
    cur = part
    for i in range(n):
        df = df.sample(frac=1)
#         train_y = df['Class'][part:]
#         features = df
#         features.drop('Class',axis = 1,inplace = True)
#         train_x = features[part:]
#         test_y = df['Class'][:part]
#         test_x = features[:part]
        train, test = train_test_split(df, test_size=part)
        part = part + cur
        k= 5
        out = knn(k,train,test)
```

```python
def knn(k,train,test):
    prediction = []
    for t in test.iterrows():
        dist= []
        for row in train.iterrows():
            s = 0
            for g in range(2,len(row)):
                s += (t[g]-row[g])**2
        dist.append([row[1],s])
        print(dist)
        arr = sorted(dist,key= lambda x: x[1])
        arr = arr[:k]
        pos,neg = 0,0
        
        for i in arr:
            if i[0]==1:
                pos+=1
            else:
                neg+=1
            
        pred = 1 if pos>neg else 0
        
        prediction.append(pred)
    return prediction
kfold(df,5)
```
